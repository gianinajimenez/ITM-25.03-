````python
## Code goes here

import csv
import re

recweekdata = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter =',')
    line_count = 0
    for row in csv_reader:
        if line_count==0:
            pass
        else:
            recweekdata.append({"first_name":row[0], "last_name":row[1], "birth_month":row[2], "birth_day":row[3],"birth_year":row[4],"course":row[5],"phone_number":row[6]})
        line_count += 1
````

````python
#first name list

first_name = list() 
counter=0
for i in recweekdata:
    first_name.append(recweekdata[counter]["first_name"])
    counter+=1
    
#last name list 

last_name = list() 
counter=0
for i in recweekdata:
    last_name.append(recweekdata[counter]["last_name"])
    counter+=1
    
#birth day list

birth_day = list() 
counter=0
for i in recweekdata:
    birth_day.append(recweekdata[counter]["birth_day"])
    counter+=1
````

````python
#cleaned birth months
#d_birth_month = dirty birth month
#c_birth_month = cleaned birth month

d_birth_month=list() 
counter=0
for i in recweekdata:
    d_birth_month.append(recweekdata[counter]['birth_month'])
    counter+=1

    unique_birth_month = set(d_birth_month)
unique_month_dict = {name:None for name in unique_birth_month}

for i in unique_month_dict:
    final_name = input(i+": "+"Enter name: ")
    unique_month_dict[i]=final_name

c_birth_month = [re.sub(r''.join(birth_month),r''.join(unique_month_dict[birth_month]),birth_month) for birth_month in d_birth_month]
c_birth_month
````

````python
#cleaned birth year

d_birth_year=list() 
c_birth_year=list()
counter=0
for i in recweekdata:
    d_birth_year.append(recweekdata[counter]['birth_year'])
    counter+=1

for i in d_birth_year:
    if len(i)==4:
        c_birth_year.append(i)
    else:
        if i[0] == 0:
            c_birth_year.append('20'+i)
        else:
            c_birth_year.append('19'+i)
c_birth_year
````

````python
#cleaned courses

d_course=list() 
counter=0
for i in recweekdata:
    d_course.append(recweekdata[counter]['course'])
    counter+=1

    unique_course = set(d_course)
unique_course_dict = {name:None for name in unique_course}

for i in unique_course_dict:
    final_name = input(i+": "+"Enter name: ")
    unique_course_dict[i]=final_name

c_course = [re.sub(r''.join(course),r''.join(unique_course_dict[course]),course) for course in d_course]
c_course
````

````python
#cleaned phone numbers

d_phone_number=list() 
c_phone_number=list()
counter=0
for i in recweekdata:
    d_phone_number.append(recweekdata[counter]['phone_number'])
    counter+=1
for i in d_phone_number:
    cleaned_phone_number = re.search(r'9.+.', i).group()
    c_phone_number.append("0"+cleaned_phone_number)

````

````python
cleaned_data=zip(first_name,last_name,c_birth_month,birth_day,c_birth_year,c_course,c_phone_number)
cleaned_data=list(cleaned_data)
````

````python
#Writing csv file of cleaned data

import csv

row = ["first name","last name","birth month","birth day","birth year","course","phone number"]

with open('cleanrecweekdata.csv', 'a') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(row)
    for i in cleaned_data:
            writer.writerow(i)
csvFile.close()
````
