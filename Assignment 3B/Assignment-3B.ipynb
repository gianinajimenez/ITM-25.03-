````python
## Code goes here
import pandas as pd
import csv
import re
import numpy as np
````

````python
#open csv

df = pd.read_csv("recweekdata.csv")
df
````

````python
#add gender

df1= pd.read_csv("studentprofiles.csv")
df["gender"] = df1["gender"]
````

````python
#add carrier

carrier = {"0917":"Globe","0918":"Smart", "0927":"Globe"}
df.loc[:,"phonenumber"].replace(r"\+63 ","0", regex=True, inplace=True)
df.loc[:,"phonenumber"].replace(r"^9","09",regex=True,inplace=True)

df["carrier"] = df.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])
````

````python
#clean birth month

unique_birthmonth = df.loc[:,"birthmonth"].unique()
two_digit_birthmonth = [0,15,19] #october, november, december

unique_birthmonth = np.delete(unique_birthmonth,two_digit_birthmonth)
other_birthmonth=['10','11','12']
````

````python
#for 2 digit birthmonths (10,11,12)
counter = 0
for i in other_birthmonth:
    other_birthmonth[counter]=input(i)
    df.loc[:,"birthmonth"].replace(i, other_birthmonth[counter],regex=True,inplace=True)
    counter+=1

#for the rest of the birth months
counter=0
for i in unique_birthmonth:
    unique_birthmonth[counter]=input(i)
    df.loc[:,"birthmonth"].replace(i, unique_birthmonth[counter],regex=True,inplace=True)
    counter+=1

df["birthmonth"]=df.loc[:,"birthmonth"].str[0:3]
````

````python
#clean birth year

df.loc[:,"birthyear"].replace(97,"1997", regex=True,inplace=True)
df.loc[:,"birthyear"].replace(98,"1998", regex=True,inplace=True)
df.loc[:,"birthyear"].replace(99,"1999", regex=True,inplace=True)
````

````python
#clean course names

unique_course = df.loc[:,"course"].unique()

counter=0
for i in unique_course:
    unique_course[counter]=input(i)
    df.loc[:,"course"].replace(r"^"+i+r"$", unique_course[counter],regex=True,inplace=True)
    counter+=1

````

````python
#write to csv

df.to_csv("cleanrecweekdata3b.csv")
````
